## Word frequency analysis
* Counting the frequency of words in a text file is an important metric in text mining.
* Create and submit a notebook that does the following:
    * Reads the supplied text file, `Sample-Text.txt` line by line
    * Gets a count of the number of times each of the words in the file is present
    * Creates a bar graph, appropriately labeled, of the twenty highest frequency words in the file.

```python
from collections import Counter
```

```python
import string
```

```python
import matplotlib.pyplot as plt
```

```python
# Read the text file
with open("/Users/vinodkumar/PycharmProjects/pythonProject3/assignment.txt", "r") as file:
    text = file.read()


# Convert the text to lowercase
text_lower = text.lower()

# Split the text into words
words = text_lower.split()

# Count the frequency of each word using Counter
word_freq = Counter(words)

# Get the twenty highest frequency words
top_words = word_freq.most_common(20)

# Extract words and their frequencies for plotting
labels, frequencies = zip(*top_words)



# Create a bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(labels, frequencies, color=plt.cm.tab20.colors)
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Top 20 Word Frequencies")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()

# Add frequency values on top of the bars
for i, freq in enumerate(frequencies):
    plt.text(i, freq, str(freq), ha='center', va='bottom')

plt.tight_layout()




# Show the plot
plt.show()
```
